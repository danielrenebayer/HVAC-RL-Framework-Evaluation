```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from RandomProcessExt import OrnsteinUhlenbeckProcess
```

```python
ou = OrnsteinUhlenbeckProcess(theta = 0.3, mu = 0.0, sigma = 0.3)
```

```python
sample_len = 500

samples = np.zeros(sample_len)
for i in range(sample_len):
    samples[i] = ou.sample()
```

```python
plt.plot(samples)
```

```python
vvars = [(0.3, 0.3), (0.2, 0.3), (0.1, 0.3),
         (0.3, 0.2), (0.2, 0.2), (0.1, 0.2),
         (0.3, 0.1), (0.2, 0.1), (0.1, 0.1)]
sample_len = 500

plt.figure(figsize=(14,8))

sequences = []
for theta, sigma in vvars:
    ou = OrnsteinUhlenbeckProcess(theta = theta, mu = 0.0, sigma = sigma)
    seq = np.zeros(sample_len)
    for idx in range(sample_len):
        seq[idx] = ou.sample()
    sequences.append(seq)

for (theta, sigma), seq in zip(vvars, sequences):
    plt.plot(seq, label=f"{theta}, {sigma}")

plt.legend()
plt.show()
```
